#### Geração da Matriz de Correlação

```python
# Gerar a matriz de correlação
corr_matrix = df.corr()
corr_matrix
```

#### **Sobre a Correlação:**

A correlação é uma medida estatística que indica como duas variáveis se movem em relação uma à outra. O coeficiente de correlação pode variar de -1 a 1, onde:

1: Indica uma correlação positiva perfeita (quando uma variável aumenta, a outra também aumenta).

0: Indica nenhuma correlação (as variáveis não têm relação linear).

-1: Indica uma correlação negativa perfeita (quando uma variável aumenta, a outra diminui).

Em resumo, df.corr() calcula a matriz de correlação para todas as colunas numéricas no DataFrame df, e essa matriz é armazenada na variável corr_matrix.

#### **Plotando a Matriz de Correlação** com Matlpotlib.Pyplot e Seaborn

```python
# Gerando a Correlação de Matriz com Matlpotlib.Pyplot e Seaborn
plt.figure(figsize=(16,16))
sns.heatmap(corr_matrix, annot=True) # O annot é de annotations que são, justamente, os valores em cada quadrado
plt.show()
```

## 3. Criando e testando o modelo com Scikit Learn

#### Dividir o df em targets (alvos) e features (características)

```python
y = df['cardio'] # Target da análise, sendo seu output
X = df.drop(columns=['cardio']) # Features da análise, sendo o input feito pelo usuário 
```

#### Conferindo o número de linhas e colunas

```python
X.shape
```

```python
y.shape
```

#### **Dividindo o dataset para treino e testes**

```python
# A função train_test_split() não somente divide o df, mas também embaralha, por padrão (shuffle=True), e distribui randomicamente os valores, para evitar víes por conta da ordem dos dados para o modelo
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) # Esse teste size representa o famoso 80% para treinamento e 20% para testes de Aprendizado Supervisionado
```

## 4. Treinando um modelo de Classificação XGBoost usando o Scikit Learn

```python
from xgboost import XGBClassifier
```

```python
xgb_classifier = XGBClassifier(objective='binary:logistic', eval_metric='error', learning_rate=0.1, max_depth=12, n_estimators=10)
xgb_classifier.fit(X_train, y_train) # Treino do modelo
```

#### Explicando os Parâmetros de XGBClassifier

- objective='binary:logistic' → Define que o problema é de classificação binária, ou seja, prever duas classes (True/False, 0/1)
  
- eval_metric='error' → Métrica de avaliação usada durante o treino. Aqui, “error” significa a taxa de erro de classificação (proporção de previsões incorretas)

- learning_rate=0.1 → Também chamado de eta. Controla o tamanho do passo em cada atualização. Valores menores tornam o aprendizado mais lento, mas podem melhorar a generalização

- max_depth=1 → Profundidade máxima das árvores de decisão. Árvores rasas (como 1) reduzem complexidade e ajudam a evitar overfitting, mas podem limitar a capacidade de capturar relações complexas

- n_estimators → Número de árvores (ou rodadas de boosting) que serão treinadas. Mais árvores aumentam a capacidade do modelo, mas também o risco de overfitting se não houver regularização


## 5. Treinando o modelo de Classificação

#### Verificando a **acurácia** do modelo

```python
result = xgb_classifier.score(X_test, y_test)
print(f'Acurácia: {result}')
```

#### Fazendo predições com os dados de teste

```python
y_predict = xgb_classifier.predict(X_test)
y_predict
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict))

# precisão (precision) é a razão TP/(TP+FP)
# revocação (recall) é a razão TP/(TP+FN)
# o F-beta score pode ser interpretado como uma média harmônica ponderada da precisão e da revocação
# onde o F-beta score atinge seu melhor valor em 1 e o pior valor em 0.
```

#### Explicação sobre possíveis resultados:

- TP (True Positive): modelo previu "positivo" e realmente era positivo.
    Exemplo: modelo disse que o paciente tem a doença, e ele realmente tem.

- FP (False Positive): modelo previu "positivo", mas na verdade era negativo.
    Exemplo: modelo disse que o paciente tem a doença, mas ele não tem (falso alarme).

- FN (False Negative): modelo previu "negativo", mas na verdade era positivo.
    Exemplo: modelo disse que o paciente não tem a doença, mas ele realmente tem (erro grave).

- TN (True Negative): modelo previu "negativo" e realmente era negativo.
    Exemplo: modelo disse que o paciente não tem a doença, e ele realmente não tem.

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, fmt = 'd', annot = True)
```
